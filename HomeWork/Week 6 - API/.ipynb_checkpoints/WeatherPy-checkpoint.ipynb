```python
#Import Modules
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
import datetime as dt
import citipy.citipy as cpy
import random
```

```python
# create empty lists
city_name = []
country_code = []
latitude = []
country =[]
date_list = []
longitude = []
Temperature = []
Humidity = []
Cloudiness = []
Wind_Speed = []
cities = []

#defining api_key, url and units to be passes to the url
api_key = 'b6907d289e10d714a6e88b30761fae22'
url = 'https://openweathermap.org/data/2.5/weather?'
units = 'imperial'

# defining the latitude and longitude ranges
lat_list = {"min":-90,"max":90}
long_list = {"min":-180,"max":180}
```

```python
#extracting the data from the json object and appending to respective lists
def data_extraction(response_json):
    latitude.append(response_json['coord']['lat'])
    longitude.append(response_json['coord']['lon'])
    Temperature.append(response_json['main']['temp_max'])
    Humidity.append(response_json['main']['humidity'])
    country.append(response_json['sys']['country'])
    date_list.append(dt.datetime.fromtimestamp(response_json['dt']))
    Cloudiness.append(response_json['wind']['speed'])
    Wind_Speed.append(response_json['clouds']['all'])
```

```python

#setting variable not_done to True
not_done = True
while not_done:
    #obtaining the nearst city from the randomly chosen latitude and longitude
    city = cpy.nearest_city(random.randint(lat_list['min'],lat_list['max']),random.randint(long_list['min'],long_list['max']))
    citi = city.city_name
    #replacing space with %20 to ensure no issues with space when calling the API
    citi = citi.replace(' ', '%20')
    #ensuring only perform further action if citi is unique and not already previously considered
    if citi not in cities:
        #generating final URL
        final_url = f"{url}&units={units}&appid={api_key}&q={citi}"
        response = requests.get(final_url)
        #checking to ensure the status of API call is a success
        if response.status_code == 200:
            #ensuring we catch any exceptions using a try catch for bad data retuned from API call while status of 200 is sent
            try:
                #converting api response to JSON data
                response_json = response.json()
                #adding unique city to the list of cities
                cities.append(citi)
                # extracting data from the json object and adding it to respective lists
                data_extraction(response_json)
                #printing url from which data is being extracted
                print(f"Processing Record {len(cities)} out of 501\n"
                f"Obtaining the weather data for city {city.city_name},{city.country_code} from:\n"
                f"{final_url}\n")
            #handling an exception bad data retuned from API call while status of 200 is sent
            except:
                print('Skipping an entry due to lack of weather data\n')
        #catching a bad response from API anything other than status 200 and skipping that data
        else:
            print("Skipping an entry due to bad response from openweathermap.org\n")
        #ensuring we stop at desired count of cities
        if len(cities)==501:
            not_done = False
            break
```

```python
#creating a dataframe with the information captured
Final_Data = pd.DataFrame(
    {
        "city":citi,
        "Country":country,
        "Date":date_list,
        "latitude":latitude,
        "longitude":longitude,
        "Max Temperature":Temperature,
        "Humidity":Humidity,
        "Cloudiness":Cloudiness,
        "Wind_Speed":Wind_Speed
    }
)
#printing the dataframe and saving it to a CSV
print(Final_Data)
Final_Data.to_csv("WeatherPy_Result.csv", index=False, header=True)
```

```python
#plotting graph 1 'City Latitudes vs. Max Temperature' and saving it to a .png file
plt.scatter(x=Final_Data['latitude'],
            y=Final_Data['Max Temperature'],
            edgecolors='k',
            alpha=0.7)
plt.xlabel('Latitudes')
plt.ylabel('Max Temperature(F)')
plt.title(f"City Latitudes vs. Max Temperature ({Final_Data['Date'].dt.date[0]})")
plt.savefig("WeatherPy_Result1.png")
plt.xlim(min(Final_Data['latitude']-10),max(Final_Data['latitude']+10))
plt.show()
```

```python
#plotting graph 3 'City Latitudes vs. Cloudiness' and saving it to a .png file
plt.scatter(x=Final_Data['latitude'],
            y=Final_Data['Cloudiness'],
            edgecolors='k',
            alpha=0.7)
plt.xlabel('Latitudes')
plt.ylabel('Cloudiness(%)')
plt.title(f"City Latitudes vs. Cloudiness ({Final_Data['Date'].dt.date[0]})")
plt.savefig("WeatherPy_Result3.png")
plt.xlim(min(Final_Data['latitude']-10),max(Final_Data['latitude']+10))
plt.show()
```

```python
#plotting graph 4 'City Latitudes vs. Wind Speed' and saving it to a .png file
plt.scatter(x=Final_Data['latitude'],
            y=Final_Data['Wind_Speed'],
            edgecolors='k',
            alpha=0.7)
plt.xlabel('Latitudes')
plt.ylabel('Wind Speed (mph)')
plt.title(f"City Latitudes vs. Wind Speed ({Final_Data['Date'].dt.date[0]})")
plt.savefig("WeatherPy_Result4.png")
plt.xlim(min(Final_Data['latitude']-10),max(Final_Data['latitude']+10))
plt.show()
```
